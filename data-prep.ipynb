# Preparing and sanitizing Datas

## Imports

```python
import pandas as pd
```

## Creating Pandas Dataframe

```python
training_df = pd.read_csv("./cs-training.csv", header=0, index_col=0)
training_df
```

## Sanitize

Using pandas.fillna() => replace NaN with the specified value

```python
training_df.fillna(0, inplace=True)
```

## Adding Features

Adding a MonthlyIncome/NumberOfDepedants

```python
training_df["IncomePerPerson"] = training_df["MonthlyIncome"] / (training_df["NumberOfDependents"] + 1)
training_df
```
